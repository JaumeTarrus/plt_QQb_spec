## Quarkonium spectrum Plots

Notebook to generate the quarkonium spectrum plots in 2207.09365 [hep-ph]. The theoretical conventional quarkonium data is from the same paper. The experimental data are all the entries from the Particle Data Group with definite $J^{PC}$. The hybrid quarkonium data is from 1510.04299 [hep-ph] and 1905.03794 [hep-ph].

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from matplotlib.lines import Line2D
```

The $J^{PC}$ of the states in the data files are labeled by a number with following assignations

$1\rightarrow 0^{-+}$,$2\rightarrow 1^{--}$, $3\rightarrow 1^{+-}$, $4\rightarrow 0^{++}$, $5\rightarrow 1^{++}$, $6\rightarrow 2^{++}$, $7\rightarrow 2^{-+}$, $8\rightarrow 2^{--}$, $9\rightarrow 3^{--}$

```python
# jpc_dic = {1 : '0^{-+}', 2 : '1^{--}', 3 : '1^{+-}', 4 : '0^{++}', 5 : '1^{++}', 6 : '2^{++}', 7 : '2^{-+}', 8 : '2^{--}', 9 :  '3^{--}'}
cc_exp = pd.read_csv('cc_exp.txt') # experimental charmonium
bb_exp = pd.read_csv('bb_exp.txt') # experimental bottomonium
cc_theo = pd.read_csv('cc_theo.txt') # theo conventional charmonium
bb_theo = pd.read_csv('bb_theo.txt') # theo conventional bottomonium
ccg1pm = pd.read_csv('ccg1pm.txt') # theo hybrid charmonium (1+- gluelump)
ccg1mm = pd.read_csv('ccg1mm.txt') # theo hybrid bottomonium (1+- gluelump)
bbg1pm = pd.read_csv('bbg1pm.txt') # theo hybrid charmonium (1-- gluelump)
bbg1mm = pd.read_csv('bbg1mm.txt') # theo hybrid bottomonium (1-- gluelump)
```

```python
# Charmonium figure

fig, ax = plt.subplots(dpi = 150)

# Experimental data

x_exp = np.array(cc_exp['jpc'])
y_exp = np.array(cc_exp['mass'])
yerr_exp = np.vstack((np.array(cc_exp['merr-']), np.array(cc_exp['merr+'])))

# Plot of the experimental data
plt.errorbar(x_exp, y_exp, yerr = yerr_exp, fmt = 'o', elinewidth = 1, capsize = 2, markersize = 3, c= 'k')
#plt.scatter(x_exp, y_exp, s = 10 , marker = 'o', c= 'k') # In case we don't want error bars for exp. data

# We create a box for each state in the theo. data. Height and width set aribitrarily. Height can be made to match theoretical
# error, however for now its too large and clutters the figure.

# Theo data for conventional quarkonium
x_theo = np.array(cc_theo['jpc'])
y_theo = np.array(cc_theo['mass'])
yerr_theo = 0.01*np.ones((2,len(y_theo))) #10**(-3)*np.vstack((np.array(cc_theo['merr']), np.array(cc_theo['merr'])))
xerr_theo = 0.3*np.ones((2,len(x_theo)))

# box creation for conventional quarkonium
errorboxes0mp = [Rectangle((x - xe[0], y - ye[0]), xe.sum(), ye.sum())
                  for x, y, xe, ye in zip(x_theo, y_theo, xerr_theo.T, yerr_theo.T)]

# Create patch collection of the boxes with specified colour/alpha
pc0mp = PatchCollection(errorboxes0mp, facecolor='r', edgecolor='none', alpha=0.3)

# Theo data for hybrid quarkonium (1+- gluelump)
x_ccg1pm = np.array(ccg1pm['jpc'])
y_ccg1pm = np.array(ccg1pm['mass'])+3.67395-4.23153+0.471 # adjusting scale to 2S
yerr_ccg1pm = 0.01*np.ones((2,len(y_ccg1pm))) #10**(-3)*np.vstack((np.array(ccg1pm['merr']), np.array(ccg1pm['merr'])))
xerr_ccg1pm = 0.3*np.ones((2,len(x_ccg1pm)))

# box creation for hybrid quarkonium (1+- gluelump)
errorboxes1pm = [Rectangle((x - xe[0], y - ye[0]), xe.sum(), ye.sum())
                  for x, y, xe, ye in zip(x_ccg1pm, y_ccg1pm, xerr_ccg1pm.T, yerr_ccg1pm.T)]

# Create patch collection with specified colour/alpha
pc1pm = PatchCollection(errorboxes1pm, facecolor='b', edgecolor='none', alpha=0.3)

# Theo data for hybrid quarkonium (1-- gluelump)
x_ccg1mm = np.array(ccg1mm['jpc'])
y_ccg1mm = np.array(ccg1mm['mass'])+3.67395-4.23153+0.471  # adjusting scale
yerr_ccg1mm = 0.01*np.ones((2,len(y_ccg1mm)))#10**(-3)*np.vstack((np.array(ccg1pm['merr']), np.array(ccg1pm['merr'])))
xerr_ccg1mm = 0.3*np.ones((2,len(x_ccg1mm)))

# box creation for hybrid quarkonium (1-- gluelump)
errorboxes1mm = [Rectangle((x - xe[0], y - ye[0]), xe.sum(), ye.sum())
                  for x, y, xe, ye in zip(x_ccg1mm, y_ccg1mm, xerr_ccg1mm.T, yerr_ccg1mm.T)]

# Create patch collection with specified colour/alpha
pc1mm = PatchCollection(errorboxes1mm, facecolor='g', edgecolor='none', alpha=0.3)


# Add patch collections to the figure 
ax.add_collection(pc0mp)
ax.add_collection(pc1pm)
#ax.add_collection(pc1mm) # hybrid quarkonium (1-- gluelump) turns out no to be relevant for the current mass range

# Plot tick lables, mass range, axes labels and title
plt.xticks(np.arange(1,10), ['$0^{-+}$','$1^{--}$','$1^{+-}$','$0^{++}$','$1^{++}$','$2^{++}$','$2^{-+}$','$2^{--}$','$3^{--}$'])
plt.ylim(2.9, 4.8)
plt.ylabel('Mass [GeV]', size = 'large')
plt.title('Charmonium spectrum')

# legend
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Experiment', markerfacecolor='k', markersize=5,),
                   Line2D([0], [0], color='r', lw= 3, alpha= 0.3 , label='Conventional'),
                   Line2D([0], [0], color='b', lw= 3, alpha= 0.3 , label='Hybrid')]
ax.legend(handles=legend_elements, loc = (0.69,0.01))

# save the figure
plt.savefig('plt_ccbar_spc.pdf', dpi = 300, format = 'pdf')

plt.show()
```
